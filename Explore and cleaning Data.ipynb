## Combine 12 months of sales data into one CSV file

```python
# Specify the directory path
directory_path = 'F:\\work in excel\\Before wok\\Sales_Data\\'

# Initialize an empty DataFrame
data = pd.DataFrame()

# List all files in the directory
files_in_directory = os.listdir(directory_path)

# Loop through each file in the directory
for file_name in files_in_directory:
    # Check if the file is a CSV file (you can customize this condition)
    if file_name.endswith('.csv'):
        # Read the CSV file into a DataFrame
        current_data = pd.read_csv(directory_path + file_name)
        
        # Concatenate the current DataFrame with the overall DataFrame
        data = pd.concat([data, current_data], ignore_index=True)

# Print the concatenated DataFrame
data
```

### Change the data type

```python
data1.info()

# Quantity Ordered => integer column
# Order ID => integer column	
# Price Each => a flaot column
# Order Date => datetime 
```

```python
# Convert 'Quantity Ordered' column to integer
data3['Quantity Ordered'] = data3['Quantity Ordered'].astype(int)

# Convert 'Order ID' column to integer
data3['Order ID'] = data3['Order ID'].astype(int)

# Convert 'Price Each' column to float
data3['Price Each'] = data3['Price Each'].astype(float)

# Convert 'Order Date' column to datetime
data3['Order Date'] = pd.to_datetime(data3['Order Date'])
```

```python
data3
```

```python
data4 = data3.copy()
```

### Extract the month, year and hour columns and Create new columns for sales and  Division of city columns

```python
# Extract the year from 'Order Date'
data4['Year'] = data4['Order Date'].dt.year

# Extract the month name from 'Order Date'
data4['Month Name'] = data4['Order Date'].dt.month_name()

# Extract the hour from 'Order Date'
data4['Hour'] = data4['Order Date'].dt.hour
```

```python
# Calculate total sales by multiplying quantity ordered and price each
data4['Sales'] = data4['Quantity Ordered'] * data4['Price Each']
```

```python
# Function to extract city from address
def extract_city(address):
    parts = address.split(',')
    city = parts[-2].strip()  # Get the second-to-last element and remove leading/trailing whitespaces
    state = parts[-1].split(" ")[-2]  # Get the last element and extract the state
    return f"{city} {state}"

# Apply the function to create the 'City' column
data4['City'] = data4['Purchase Address'].apply(extract_city)
```

### Create a category column for price Each

```python
data4['Price Each'].describe()
```

```python
# Create a 'Price Category' column based on price ranges
data4['Price Category'] = pd.cut(data4['Price Each'], 
                                 bins=[0, 20, 250, 750, float('inf')], 
                                 labels=['Low', 'Medium', 'High', 'Very High'])

# Display the DataFrame with the new 'Price Category' column
print(data4[['Price Each', 'Price Category']])
```

```python
# Data frame arrangement
selected_columns = ['Order ID', 'Product', 'Quantity Ordered', 'Price Each', 'Sales', 'Price Category', 'Order Date', 'Year', 'Month Name', 'Hour', 'Purchase Address', 'City']

# Create a new DataFrame with the selected columns
data5 = data4[selected_columns]
data5
```
